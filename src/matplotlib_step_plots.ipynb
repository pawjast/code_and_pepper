# `matplotlib` step plots

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Data

```python
x = [1, 4, 10, 12, 15, 20]
y = [2, 4, 1, 3, 1, 4]
```

## ax.step()

```python
fig, axs = plt.subplots(
    3, 1,
    figsize=(16,12),
    facecolor="white",
    sharex=True,
    sharey=True
)
fig.suptitle(
    "ax.step(x,y)",
    fontsize=20,
)

# plot points
for ax in axs:
    ax.scatter(
        x, y,
        marker="X",
        s=275,
        linewidth=1.5,
        edgecolor="black",
        facecolor="white",
        zorder=100  # crude way to keep the points always on top
    )

# default step: `pre`
axs[0].step(
    x, y,
    color="red",
    label="ax.step(x, y)"
)
axs[1].step(
    x, y,
    color="blue",
    where="mid",
    label="ax.step(x, y, where='mid')"
)
axs[2].step(
    x, y,
    color="green",
    where="post",
    label="ax.step(x, y, where='post')"
)

# set axis limits
axs[0].set_ylim(0, 5.5)
axs[0].set_yticks(range(1, 6))
axs[0].set_xlim(0, 21)
axs[2].set_xticks(range(1, 21))

# add legend
for ax in axs:
    ax.legend(
        fontsize=16,
        loc="upper center",
        facecolor="whitesmoke",
        edgecolor="none"
    )

    # Update font size of the 'x' and 'y' axis labels
    ax.tick_params(
        axis="both",
        labelsize=16
    )

    ax.grid(
        color="gainsboro",
        linewidth=0.5
    )

# Remove spines
for position in ["top", "right"]:
    for ax in axs:
        ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```

## ax.plot()

```python
fig, axs = plt.subplots(
    3, 1,
    figsize=(16,12),
    facecolor="white",
    sharex=True,
    sharey=True
)
fig.suptitle(
    "ax.plot(y, drawstyle=...)",
    fontsize=20,
)

# plot points
for ax in axs:
    ax.scatter(
        range(len(y)), y,
        marker="X",
        s=275,
        linewidth=1.5,
        edgecolor="black",
        facecolor="white",
        zorder=100  # crude way to keep the points always on top
    )

axs[0].plot(
    y,
    color="red",
    drawstyle="steps-pre",
    label="ax.plot(y, drawstyle='steps-pre')"
)
axs[1].plot(
    y,
    color="blue",
    drawstyle="steps-mid",
    label="ax.plot(y, drawstyle='steps-mid')"
)
axs[2].plot(
    y,
    color="green",
    drawstyle="steps-post",
    label="ax.plot(y, drawstyle='steps-post')"
)

# set axis limits
axs[0].set_ylim(0, 5.5)
axs[0].set_yticks(range(1, 6))
axs[0].set_xlim(-1, 6)
axs[2].set_xticks(range(0, 6))

# add legend
for ax in axs:
    ax.legend(
        fontsize=16,
        loc="upper center",
        facecolor="whitesmoke",
        edgecolor="none"
    )

    # Update font size of the 'x' and 'y' axis labels
    ax.tick_params(
        axis="both",
        labelsize=16
    )

    ax.grid(
        color="gainsboro",
        linewidth=0.5
    )

# Remove spines
for position in ["top", "right"]:
    for ax in axs:
        ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot2.png",
    format="png",
    dpi=125
);
```

## ax.plot() & ax.fill_between()

```python
fig, axs = plt.subplots(
    3, 1,
    figsize=(16,12),
    facecolor="white",
    sharex=True,
    sharey=True
)
fig.suptitle(
    "ax.plot(y, drawstyle=...) & ax.fill_between(..., steps=...)",
    fontsize=20,
)

# plot points
for ax in axs:
    ax.scatter(
        range(len(y)), y,
        marker="X",
        s=275,
        linewidth=1.5,
        edgecolor="black",
        facecolor="white",
        zorder=100  # crude way to keep the points always on top
    )

axs[0].plot(
    y,
    color="red",
    drawstyle="steps-pre",
    label="ax.plot(y, drawstyle='steps-pre')"
)
axs[0].fill_between(
    range(len(y)), y,
    step="pre",
    color="red",
    alpha=0.1
)
axs[1].plot(
    y,
    color="blue",
    drawstyle="steps-mid",
    label="ax.plot(y, drawstyle='steps-mid')"
)
axs[1].fill_between(
    range(len(y)), y,
    step="mid",
    color="blue",
    alpha=0.1
)
axs[2].plot(
    y,
    color="green",
    drawstyle="steps-post",
    label="ax.plot(y, drawstyle='steps-post')"
)
axs[2].fill_between(
    range(len(y)), y,
    step="post",
    color="green",
    alpha=0.1
)

# set axis limits
axs[0].set_ylim(0, 5.5)
axs[0].set_yticks(range(1, 6))
axs[0].set_xlim(-1, 6)
axs[2].set_xticks(range(0, 6))

# add legend
for ax in axs:
    ax.legend(
        fontsize=16,
        loc="upper center",
        facecolor="whitesmoke",
        edgecolor="none"
    )

    # Update font size of the 'x' and 'y' axis labels
    ax.tick_params(
        axis="both",
        labelsize=16
    )

    ax.grid(
        color="gainsboro",
        linewidth=0.5
    )

# Remove spines
for position in ["top", "right"]:
    for ax in axs:
        ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot3.png",
    format="png",
    dpi=125
);
```

## ax.stairs()

```python
fig, axs = plt.subplots(
    2, 1,
    figsize=(16,12),
    facecolor="white",
    sharex=True,
    sharey=True
)
fig.suptitle(
    "ax.stairs()",
    fontsize=20,
)

# plot points
for ax in axs:
    ax.scatter(
        x, y,
        marker="X",
        s=275,
        linewidth=1.5,
        edgecolor="black",
        facecolor="white",
        zorder=100  # crude way to keep the points always on top
    )

axs[0].stairs(
    values=y,  # step height
    edges=[x[0], *x],  # edge width
    baseline=None,
    color="red",
    linewidth=1.5,
    label="steps-pre: duplicate 1st element in `edges=[x[0], *x]`"
)
axs[1].stairs(
    values=y,  # step height
    edges=[*x, x[-1]],  # edge
    baseline=None,
    color="green",
    linewidth=1.5,
    label="steps-post: duplicate last element in `edges=[*x, x[-1]]`"
)

# set axis limits
axs[0].set_ylim(0, 5.5)
axs[0].set_yticks(range(1, 6))
axs[0].set_xlim(0, 21)
axs[1].set_xticks(range(1, 21))

# add legend
for ax in axs:
    ax.legend(
        fontsize=16,
        loc="upper center",
        facecolor="whitesmoke",
        edgecolor="none"
    )

    # Update font size of the 'x' and 'y' axis labels
    ax.tick_params(
        axis="both",
        labelsize=16
    )

    ax.grid(
        color="gainsboro",
        linewidth=0.5
    )

# Remove spines
for position in ["top", "right"]:
    for ax in axs:
        ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot4.png",
    format="png",
    dpi=125
);
```
