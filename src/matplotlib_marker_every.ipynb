# `matplotlib` plot with data marker every # point

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Data

```python
step = 0.0025
x = np.arange(-2.8, 2.8+step, step)
y = 0.25*x**2 + 0.5*x*np.sin(20*x) + 0.5
```

## Plot

```python
line_width_val = 2
colour_red = "#c01c28"  # GNOME red 4
colour_green = "#2ec27e"  # GNOME green 4
colour_blue = "#1c71d8"  # GNOME blue 4

fig, axs = plt.subplots(
    1, 2,
    figsize=(10,6),
    facecolor="white",  # GNOME red 1
)
fig.suptitle(
    "Control density of data markers",
    fontsize=20,
)

# Plot 1 - default
axs[0].plot(
    x, y,
    marker="o",
    markersize=5,
    color=colour_blue,
    markerfacecolor=colour_blue,
    markeredgecolor="none"
)
axs[0].set_title(
    "Default",
    fontsize=16,
    color=colour_red
    )

# Plot 2 - less dense markers
axs[1].plot(
    x, y,
    marker="o",
    markersize=5,
    color=colour_blue,
    markerfacecolor=colour_blue,
    markeredgecolor="none",
    markevery=25
)
axs[1].set_title(
    "with `markevery=` param",
    fontsize=16,
    color=colour_green
    )

for ax in axs:

    # Set axis limit
    ax.set_xlim(-3, 3)
    ax.set_ylim(0, 4)

    # Remove `y` axis ticks and tick labels
    ax.tick_params(
        axis="both",
        which="major",
        left=False,
        labelleft=False,
        bottom=False,
        labelbottom=False
    )

    # Update spines thickness
    for position in ["left", "bottom"]:
        ax.spines[position].set_linewidth(line_width_val)
        # set colour
        if ax == axs[0]:
            ax.spines[position].set_color(colour_red)
        else:
            ax.spines[position].set_color(colour_green)


    # Move "y" spine to the center
    ax.spines["left"].set_position("center")

    # Remove spines
    for position in ["top", "right"]:
        ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```
