# `matplotlib` r-style bar chart

## Imports

```python
import matplotlib.pyplot as plt
```

## Data

```python
x = list("ABCDE")
y = [3, 5, 1, 4, 2]
```

## Plot

```python
# Settings
line_width_val = 1.5
label_font_size=18
```

### Grey backgorund

```python
fig, ax = plt.subplots(
    figsize=(10, 8),
    facecolor="white"
)

ax.set_facecolor("#f6f5f4")

fig.suptitle(
    "R-style bar chart",
    fontsize=25,
    fontfamily="serif",
    weight="bold"
)

ax.bar(
    x=x,
    height=y,
    color="white",
    edgecolor="black",
    linewidth=line_width_val,
    zorder=10
)

ax.set_xlabel(
    "Categories",
    fontsize=label_font_size
)
ax.set_ylabel(
    "Count",
    fontsize=label_font_size
)
ax.set_ylim(0, 5)
ax.tick_params(
    axis="x",
    length=0,
    labelsize=label_font_size
)
ax.tick_params(
    axis="y",
    length=10,
    labelsize=label_font_size
)

# Offset visible spines
for position in ["left", "bottom"]:
    ax.spines[position].set_position(
        ("outward", 10)
    )

# Hide some spines
for position in ["top", "right", "bottom"]:
    ax.spines[position].set_visible(False)


ax.grid(
    color="white",
    linewidth=line_width_val
)

# Footer
ax.text(
    1, -0.15,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

plt.tight_layout()

# plt.savefig(
#     fname="1.png",
#     format="png",
#     dpi=125
# );
```
