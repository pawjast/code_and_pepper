## Data

```python
df = pd.DataFrame(
    [3, 7, 5, 4],
    columns=["A"],
    index=["2023-01-01", "2023-02-01", "2023-03-01", "2023-04-01"]
)

df
```

```python
df1 = pd.DataFrame(
    [3, 7, 5, 4],
    columns=["A"],
    index=pd.date_range(
        start='01/01/2023',
        periods=4,
        freq="MS"  # month start
    )
)

df1
```

## Plot df
`Index` are dates but stored as strings so matplotlib doesn't know these are dates.

```python
fig, ax = plt.subplots(
    figsize=(6, 6),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "dates as strings",
    fontsize=22,
    weight="bold"
)

ax.bar(
    x=df.index,
    height=df["A"],
    color="#3584e4"
)
ax.set_axisbelow(True)  # move grid behind the bars

# Footer
ax.text(
    1, -0.17,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Plot df1
The one with `Index` using the `DateTime` type.

```python
fig, ax = plt.subplots(
    figsize=(6, 6),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "dates as dates",
    fontsize=22,
    weight="bold"
)

ax.bar(
    x=df1.index,
    height=df1["A"],
    color="#3584e4"
)
ax.set_axisbelow(True)  # move grid behind the bars

# Footer
ax.text(
    1, -0.17,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Plot df1 with updated width
Since we're plotting data with monthly frequency, we should update the bar width to something like 25 days

```python
fig, ax = plt.subplots(
    figsize=(6,6),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "date as dates with proper width",
    fontsize=22,
    weight="bold"
)

ax.bar(
    x=df1.index,
    height=df1["A"],
    width=25,
    color="#3584e4"
)
ax.set_axisbelow(True)  # move grid behind the bars
ax.set_xticks(df1.index)

# Footer
ax.text(
    1, -0.17,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
