## Data

```python
data = [1, 5, 5, 3, 6, 2, 1, 3, 7, 7, 5]
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 24,
        "hatch.linewidth": 1.75,
        "axes.spines.top": False,
        "axes.spines.right": False,
        # "axes.spines.left": False,
        # "axes.spines.bottom": False,
    }
)
```

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white",
    layout="constrained"
)
ax.set_title(
    "Minimalistic line plot",
    fontsize=40,
    weight="bold",
    loc="left",
    y=1.02
)

ax.plot(
    data,
    lw=2.5,
    marker="o",
    markersize=16,
    markerfacecolor="black",
    markeredgecolor="white",
    markeredgewidth=6,
    color="black"
)

ax.set_xlim(-0.5, 10.5)
ax.set_ylim(0, 8)

# Footer
ax.text(
    1, -0.1,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=18,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Plot function

```python
def my_plot(data, color):
    _, ax = plt.subplots(
        figsize=(16,10),
        facecolor="white",
        layout="constrained"
    )
    ax.set_title(
        "Minimalistic line plot",
        color=color,
        fontsize=40,
        weight="bold",
        loc="left",
        y=1.02
    )

    ax.plot(
        data,
        lw=2.5,
        marker="o",
        markersize=16,
        markerfacecolor=color,
        markeredgecolor="white",
        markeredgewidth=6,
        color=color
    )

    ax.set_xlim(-0.5, 10.5)
    ax.set_ylim(0, 8)

    # Footer
    ax.text(
        1, -0.12,
        "by @pawjast\nmade with @matplotlib \u2665",
        transform = ax.transAxes,
        horizontalalignment="right",
        verticalalignment="bottom",
        color="#9a9996",
        fontsize=18,
        family="monospace",
        weight="medium"
    )

    return ax
```

```python
my_plot(
    data=data,
    color="red"
)
```

```python
my_plot(
    data=data,
    color="green"
)
```
