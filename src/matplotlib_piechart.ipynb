## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
        "axes.spines.top": False,
        "axes.spines.right": False,
        # "axes.spines.left": False,
        # "axes.spines.bottom": False,
    }
)
```

```python
im = plt.imread("python.png")
```

```python
with plt.xkcd():
    fig, ax = plt.subplots(
        figsize=(16,12),
        facecolor=("white", 0.5),
        layout="constrained"
    )
    fig.suptitle(
        "Does it still make sense to learn Python in 2025?",
        fontsize=40,
        weight="bold"
    )

    ax.imshow(im, alpha=0.25)
    ax.axis("off")

    rect1 = 0.0, 0.0, 0.96, 0.90 #  x, y, width, heigh
    ax1 = fig.add_axes(rect1)

    ax1.pie(
        x=data.values(),
        labels=data.keys(),
        colors=[
            "#3e7cf7",  # blue
            "#fcf92e"  # yellow
        ],
        startangle=90,
        wedgeprops=dict(
            linewidth=6,
            edgecolor="#4e4e4a"
        ),
        textprops=dict(
            size=30
        ),
        hatch=[".", "x"]
    )

    # Footer
    ax.text(
        1, 0,
        "by @pawjast\nmade with @matplotlib \u2665",
        transform = ax.transAxes,
        horizontalalignment="right",
        verticalalignment="bottom",
        color="#9a9996",
        fontsize=16,
        family="monospace",
        weight="medium"
    )

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
