## Data

```python
df = pd.DataFrame.from_dict(
    {
        "A": 9,
        "B": 4,
        "C": 1,
        "D": 4,
        "F": 9,
    },
    orient="index",
    columns=["data"]
)

df
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
        "hatch.color": "black",
        
    }
)
```

### Access spines: way 1

Using `for` loop. This is what I used to use most of the time.

```python
fig, ax = plt.subplots(
    figsize=(16,9),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "I wanna be a parabola",
    fontsize=25,
    weight="bold"
)

ax.bar(
    df.index,
    df["data"],
    facecolor=(1, 1, 1, 0),
    hatch=".."
)

for position in ["left", "right", "top"]:
    ax.spines[position].set_visible(False)

# Footer
ax.text(
    1, -.12,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

### Access spines: way 2

Using `list` of spine names. This is more concise way than `for` loop.

```python
fig, ax = plt.subplots(
    figsize=(16,9),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "I wanna be a parabola",
    fontsize=25,
    weight="bold"
)

ax.bar(
    df.index,
    df["data"],
    facecolor=(1, 1, 1, 0),
    hatch=".."
)

ax.spines[["left", "right", "top"]].set_visible(False)

# Footer
ax.text(
    1, -.12,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

### Access spines: way 3

Using `.<spine_name>` to access individual spines.

```python
fig, ax = plt.subplots(
    figsize=(16,9),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "I wanna be a parabola",
    fontsize=25,
    weight="bold"
)

ax.bar(
    df.index,
    df["data"],
    facecolor=(1, 1, 1, 0),
    hatch=".."
)


ax.spines.left.set_visible(False)
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)

# Footer
ax.text(
    1, -.12,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
