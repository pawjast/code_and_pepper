# `matplotlib` minor ticks

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Plot: default

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white"
)
fig.suptitle(
    "Default plot",
    fontsize=24,
)

# set axis limits
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# set major ticks: 
ax.set_xticks(
    [0, 1, 2, 3]
)
ax.set_yticks(
    [0, 1, 2, 3]
)

# add grid: by default show only major grid lines on both axes
ax.grid(
    color="blue"
)

# font size for tick labels
ax.tick_params(
    labelsize=16
)

plt.tight_layout()

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```

```python
# Major ticks
ax.get_xticks()
```

```python
# There are do minor ticks at the moment at the x-axis
ax.get_xticks(minor=True)
```

## Plot 2: with minor ticks

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white"
)
fig.suptitle(
    "Minor ticks",
    fontsize=24,
)

# set axis limits
ax.set_xlim(0, 3)
ax.set_ylim(0, 3)

# set major ticks: 
ax.set_xticks(
    [0, 1, 2, 3]
)
ax.set_yticks(
    [0, 1, 2, 3]
)

# add grid: by default show only major grid lines on both axes
ax.grid(
    color="blue",
    linewidth=2
)

# update major tick appearance
ax.tick_params(
    which="major",
    length=5,
    color="blue",
    width=2
)

# set minor ticks
ax.set_xticks(
    [0.5, 1.33, 1.66, 2.25, 2.5, 2.75],
    minor=True
)
ax.set_yticks(
    [0.5, 1.33, 1.66, 2.25, 2.5, 2.75],
    minor=True
)

# display minor grid
ax.grid(
    which="minor",
    color="red",
    linewidth=1,
    linestyle="-."
)

# update minor tick appearance
ax.tick_params(
    which="minor",
    length=10,
    color="red",
    width=1
)

# font size for tick labels
ax.tick_params(
    labelsize=16
)

plt.tight_layout()

plt.savefig(
    fname="plot2b.png",
    format="png",
    dpi=125
);
```

```python
# Major ticks
ax.get_xticks()
```

```python
# Minor ticks
ax.get_xticks(minor=True)
```
