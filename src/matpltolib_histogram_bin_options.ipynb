# `matplotlib` histogram bin options

## Imports

```python
import random

import matplotlib.pyplot as plt
```

## Data

```python
sample_size = 3000

data = [
    random.normalvariate(mu=0,sigma=1)
    for _ in range(sample_size)
]
```

## Plot

```python
grey_col = "#3d3846"  # GNOME dark 3
```

### Custom number of bins

```python
fig, ax = plt.subplots(
    figsize=(14,16),
    facecolor="white"
)
fig.suptitle(
    "`bins=12`",
    fontsize=35,
    fontfamily="serif",
    weight="heavy",
    color=grey_col
)

ax.hist(
    data,
    bins=12,
    edgecolor="#26a269",  # GNOME green 5
    facecolor ="#57e389"  # GNOME green 2
)

# labels
ax.set_xlabel (
    "variable",
    fontsize=20,
    color=grey_col
)
ax.set_ylabel(
    "count",
    fontsize=20,
    color=grey_col
)

# Update 'x' and 'y' axis tick labels size and colour
ax.tick_params(
    axis="both",
    labelsize=20,
    color=grey_col,  # tick colour
    labelcolor=grey_col
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

# Remove spines
for position in ["bottom", "left"]:
    ax.spines[position].set_color(grey_col)

plt.tight_layout()

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```

### Specific bins boundaries

```python
fig, ax = plt.subplots(
    figsize=(14,16),
    facecolor="white"
)
fig.suptitle(
    "`bins=[-4.5, -3, -1.5, 0, 1.5, 3, 4.5]`",
    fontsize=35,
    fontfamily="serif",
    weight="heavy",
    color=grey_col
)

ax.hist(
    data,
    bins=[-4.5, -3, -1.5, 0, 1.5, 3, 4.5],
    edgecolor="#c64600",  # GNOME orange 5
    facecolor ="#ffa348"  # GNOME orange 2
)

# labels
ax.set_xlabel (
    "variable",
    fontsize=20,
    color=grey_col
)
ax.set_ylabel(
    "count",
    fontsize=20,
    color=grey_col
)

# Update 'x' and 'y' axis tick labels size and colour
ax.tick_params(
    axis="both",
    labelsize=20,
    color=grey_col,  # tick colour
    labelcolor=grey_col
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

# Remove spines
for position in ["bottom", "left"]:
    ax.spines[position].set_color(grey_col)

plt.tight_layout()

plt.savefig(
    fname="plot2.png",
    format="png",
    dpi=125
);
```
