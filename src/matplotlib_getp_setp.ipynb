# `matplotlib` using `.setp` and `.getp`

## Imports

```python
import matplotlib.pyplot as plt
```

```python
from matplotlib import font_manager

# Add system fonts so matplotlib sees them
for font in font_manager.findSystemFonts(fontpaths=None, fontext='ttf'):
    font_manager.fontManager.addfont(font)
```

## Data

```python
x = [6, 2, 5, 3, 4, -10]
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
    }
)
```

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "My motivation throughout the day",
    fontsize=25,
    weight="bold"
)

line, = ax.plot(
    x,
    label="Data"
)

lgd = ax.legend()
ax.grid()


# Footer
ax.text(
    1, -0.08,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Look up properties

```python
plt.getp(fig)
```

```python
plt.getp(ax)
```

```python
plt.getp(line)
```

```python
plt.getp(lgd)
```

## Set up new properties

```python
plt.setp(
    fig,
    facecolor="azure"
)

plt.setp(
    ax,
    frame_on=False,
    ylim = (-12, 12)
)

plt.setp(
    line,
    linestyle="dashed",
    linewidth=4,
    color="dodgerblue"
)

plt.setp(
    lgd,
    title="The legend",
)
```
