# GNOME colour palette

[reference](https://developer.gnome.org/hig/reference/palette.html)

## Imports

```python
import matplotlib.pyplot as plt
```

## Data

All GNOME colouts in HEX

```python
blue = {
    "blue 1": "#99c1f1",
    "blue 2": "#62a0ea",
    "blue 3": "#3584e4",
    "blue 4": "#1c71d8",
    "blue 5": "#1a5fb4",
}

green = {
    "green 1": "#8ff0a4",
    "green 2": "#57e389",
    "green 3": "#33d17a",
    "green 4": "#2ec27e",
    "green 5": "#26a269",
}

yellow = {
    "yellow 1": "#f9f06b",
    "yellow 2": "#f8e45c",
    "yellow 3": "#f6d32d",
    "yellow 4": "#f5c211",
    "yellow 5": "#e5a50a",
}

orange = {
    "orange 1": "#ffbe6f",
    "orange 2": "#ffa348",
    "orange 3": "#ff7800",
    "orange 4": "#e66100",
    "orange 5": "#c64600",
}

red = {
    "red 1": "#f66151",
    "red 2": "#ed333b",
    "red 3": "#e01b24",
    "red 4": "#c01c28",
    "red 5": "#a51d2d",
}

purple = {
    "purple 1": "#dc8add",
    "purple 2": "#c061cb",
    "purple 3": "#9141ac",
    "purple 4": "#813d9c",
    "purple 5": "#613583",
}

brown = {
    "brown 1": "#cdab8f",
    "brown 2": "#b5835a",
    "brown 3": "#986a44",
    "brown 4": "#865e3c",
    "brown 5": "#63452c",
}

light = {
    "light 1": "#ffffff",
    "light 2": "#f6f5f4",
    "light 3": "#deddda",
    "light 4": "#c0bfbc",
    "light 5": "#9a9996",
}

dark = {
    "dark 1": "#77767b",
    "dark 2": "#5e5c64",
    "dark 3": "#3d3846",
    "dark 4": "#241f31",
    "dark 5": "#000000",
}

all_colours = [
    blue,
    green,
    orange,
    red,
    purple,
    brown,
    light,
    dark
]
```

## Plot

```python
fig, axs = plt.subplots(
    len(all_colours), 5, 
    figsize=(10, 5),
    facecolor="white"
)
fig.suptitle(
    "GNOME colours",
    fontsize=20,
    fontweight="book"
)

for ax, colour in zip(axs, all_colours):
    for a, key in zip(ax, colour):
        a.set_facecolor(colour[key])
        a.tick_params(
            labelsize=0,
            length=0
        )
        if "light" in key:
            c = "black"
        else:
            c = "white"

        a.text(
            0.5, 0.5,
            key,
            fontsize=14,
            fontweight="demibold",
            color=c,
            ha="center",
            va="center_baseline"
        )

        for position in ["top", "right", "left", "bottom"]:
            a.spines[position].set_visible(False)


plt.tight_layout()
plt.subplots_adjust(wspace=0.025, hspace=0.075)

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```
