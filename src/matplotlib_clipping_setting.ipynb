## Data

```python
x_vals = [1, 2, 3.95, 0]
y_vals = [1, 2.97, 0, 2.5]
```

## Plot with clipping (default)

* bottom and right gridline is clipped (visible when spines are removed)
* points at the edge of the canvas are clipped too

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
    }
)
```

```python
fig, ax = plt.subplots(
    figsize=(16,9),
    layout="constrained"
)
fig.suptitle(
    "`clip_on` set to True (default)",
    fontsize=40,
    weight="bold"
)

# Remove bg colour and spines
ax.set_frame_on(False)

SIZE = 4

ax.set_xticks(
    ticks=range(0, SIZE+1)
)
ax.set_yticks(
    ticks=range(0, SIZE)
)
ax.set_xlim(0, SIZE)
ax.set_ylim(0, SIZE-1)

# Set ticks colour to invisible
ax.tick_params(
    color=(0, 0, 0, 0)
)

ax.grid(
    color="dimgray",
    lw=1
)

# Plot
ax.scatter(
    x=x_vals,
    y=y_vals,
    s=2000,
    marker="o",
    c="lightcyan",
    edgecolors="blue",
    linewidths=2,
    zorder=10
)

# Footer
ax.text(
    1, -0.1,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Plot with no clipping

```python
fig, ax = plt.subplots(
    figsize=(16,9),
    layout="constrained"
)
fig.suptitle(
    "`clip_on` set to False",
    fontsize=40,
    weight="bold"
)

# Remove bg colour and spines
ax.set_frame_on(False)

SIZE = 4

ax.set_xticks(
    ticks=range(0, SIZE+1)
)
ax.set_yticks(
    ticks=range(0, SIZE)
)
ax.set_xlim(0, SIZE)
ax.set_ylim(0, SIZE-1)

# Set ticks colour to invisible
ax.tick_params(
    color=(0, 0, 0, 0)
)

ax.grid(
    color="dimgray",
    lw=1,
    clip_on=False
)

# Plot
ax.scatter(
    x=x_vals,
    y=y_vals,
    s=2000,
    marker="o",
    c="lightcyan",
    edgecolors="blue",
    linewidths=2,
    zorder=10,
    clip_on=False
)

# Footer
ax.text(
    1, -0.1,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
