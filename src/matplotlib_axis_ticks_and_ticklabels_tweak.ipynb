# Tweak `matplotlib` axis ticks and tick labels

## Imports

```python
import matplotlib.pyplot as plt
```

## Default plot

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor="#ffbe6f"
)
fig.suptitle(
    "Default plot 1",
    fontsize=16
)

ax.plot(
    [0.20, 0.60, 0.40, 0.90, 0.30, 0.50],
    color="black",
    marker="o"
)

plt.tight_layout()

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```

## Tweaked plot 1

* Add *axis limits*
* use slicing to remove **1st** and **last** tick
* use **`f-strings`** to format the values the way you want

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor="#8ff0a4"
)
fig.suptitle(
    "Tweaked plot 1",
    fontsize=16
)

ax.plot(
    [0.20, 0.60, 0.40, 0.80, 0.30, 0.50],
    color="black",
    marker="o"
)

# Set axis limits
ax.set_xlim(
    [-1, 6]
)
ax.set_ylim(
    [0, 1.25]
)

# Specify `x` and `y` axis values
y_vals = [0, 0.25, 0.50, 0.75, 1.00, 1.25]
x_vals = [-1, 0, 1, 2, 3, 4, 5, 6]

# Set ticks and tick labels
ax.set_yticks(
    y_vals[1:-1]
)
ax.set_yticklabels(
    [f"{y:.2f}" for y in y_vals[1:-1]]
)
ax.set_xticks(
    x_vals[1:-1]
)
# Use `f-strings` to tweak the appearance of the date
ax.set_xticklabels(
    [f"{x:d}" for x in x_vals[1:-1]]
)

plt.tight_layout()

plt.savefig(
    fname="plot2.png",
    format="png",
    dpi=125
);
```

**Note:** there are multiple ways to create values for `x_vals` and `y_vals`:

* `range()` - [doc](https://docs.python.org/3/library/stdtypes.html#range)
* `np.arange()` - [doc](https://numpy.org/doc/stable/reference/generated/numpy.arange.html)
* `np.linspace()` - [doc](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html)
* `pd.date_range()` - [doc](https://pandas.pydata.org/docs/reference/api/pandas.date_range.html)

## Tweaked plot 2 - "x" as date

* use `f-strings` to format the values the way you want

```python
import pandas as pd
```

### Default date

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor="#ffbe6f"
)
fig.suptitle(
    "Default plot 2",
    fontsize=16
)

# define dates for x values
x_vals = pd.date_range(
    start="1/4/2023",
    periods=len(y_vals)
)

ax.plot(
    x_vals,
    [0.20, 0.60, 0.40, 0.80, 0.30, 0.50],
    color="black",
    marker="o"
)

# Set axis limits
ax.set_ylim(
    [0, 1.25]
)

# Specify `x` and `y` axis values
y_vals = [0, 0.25, 0.50, 0.75, 1.00, 1.25]

# Set ticks and tick labels
ax.set_yticks(
    y_vals[1:-1]
)
# Use `f-strings` to tweak the appearance of the x values
ax.set_yticklabels(
    [f"{y:.2f}" for y in y_vals[1:-1]]
)
ax.set_xticks(
    x_vals
)

plt.tight_layout()

plt.savefig(
    fname="plot3.png",
    format="png",
    dpi=125
);
```

### Tweaked date

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor="#8ff0a4"
)
fig.suptitle(
    "Tweaked plot 2",
    fontsize=16
)

# define dates for x values
x_vals = pd.date_range(
    start="1/4/2023",
    periods=len(y_vals)
)

ax.plot(
    x_vals,
    [0.20, 0.60, 0.40, 0.80, 0.30, 0.50],
    color="black",
    marker="o"
)

# Set axis limits
ax.set_ylim(
    [0, 1.25]
)

# Specify `x` and `y` axis values
y_vals = [0, 0.25, 0.50, 0.75, 1.00, 1.25]

# Set ticks and tick labels
ax.set_yticks(
    y_vals[1:-1]
)
# Use `f-strings` to tweak the appearance of the y values
ax.set_yticklabels(
    [f"{y:.2e}" for y in y_vals[1:-1]]
)
ax.set_xticks(
    x_vals
)
# Use `f-strings` to tweak the appearance of the x values (dates)
ax.set_xticklabels(
    [f"{x:%A}" for x in x_vals]
)

plt.tight_layout()

plt.savefig(
    fname="plot4.png",
    format="png",
    dpi=125
);
```
