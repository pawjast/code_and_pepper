## Data

```python
max_x = 18
max_y = 12

random.seed(0)
x = sorted(random.sample(range(1, max_x), 6))
random.seed(1)
y = sorted(random.sample(range(1, max_y), 6))

print(f"x: {x}\ny: {y}")
```

## Plot 1 - default

```python
# Settings

my_colour = "#26a269"
grid_colour = "#3584e4"

plt.rcParams.update(
    {
        "figure.figsize": (12, 8),
        "font.family": "Roboto",
        "font.size": 22,
    }
)
```

```python
fig, ax = plt.subplots(
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "Step data - default appearance",
    fontsize=25,
    weight="bold"
)

ax.step(
    x, y,
    color=my_colour,
    lw=1.75,
    where="post",
    zorder=1
)
ax.scatter(
    x, y,
    marker="o",
    facecolor="white",
    edgecolors=my_colour,
    linewidths=1.75,
    zorder=2
)

# Remove spines
for position in["top", "right"]:
    ax.spines[position].set_visible(False)

ax.set_xlim(0, max_x+1)
ax.set_ylim(0, max_y+1)

# Footer
ax.text(
    1, -0.15,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

ax.grid(
    color=grid_colour,
    lw=0.15,
    ls=(0, (45, 15))
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Plot 2: with tweaked ticks and labels

```python
fig, ax = plt.subplots(
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "Step data - with custom ticks",
    fontsize=25,
    weight="bold"
)

ax.step(
    x, y,
    color=my_colour,
    lw=1.75,
    where="post",
    zorder=1
)
ax.scatter(
    x, y,
    marker="o",
    facecolor="white",
    edgecolors=my_colour,
    linewidths=1.75,
    zorder=2
)

# Remove spines
for position in["top", "right"]:
    ax.spines[position].set_visible(False)

ax.set_xlim(0, max_x+1)
ax.set_ylim(0, max_y+1)

ax.set_xticks(x)
ax.set_xticklabels(x)
ax.set_yticks(y)
ax.set_yticklabels([f"{val}°" for val in y])

# Footer
ax.text(
    1, -0.15,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

ax.grid(
    color=grid_colour,
    lw=0.15,
    ls=(0, (45, 15))
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Plot 3: with tweaked ticks and labels in one go

```python
fig, ax = plt.subplots(
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "Step data - with custom ticks",
    fontsize=25,
    weight="bold"
)

ax.step(
    x, y,
    color=my_colour,
    lw=1.75,
    where="post",
    zorder=1
)
ax.scatter(
    x, y,
    marker="o",
    facecolor="white",
    edgecolors=my_colour,
    linewidths=1.75,
    zorder=2
)

# Remove spines
for position in["top", "right"]:
    ax.spines[position].set_visible(False)

ax.set_xlim(0, max_x+1)
ax.set_ylim(0, max_y+1)

ax.set_xticks(
    ticks=x,
    labels=x
)
ax.set_yticks(
    ticks=y,
    labels=[f"{val}°" for val in y]
)

# Footer
ax.text(
    1, -0.15,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

ax.grid(
    color=grid_colour,
    lw=0.15,
    ls=(0, (45, 15))
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
