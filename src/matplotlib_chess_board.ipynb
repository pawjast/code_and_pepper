## Data

```python
blacks = np.array([bool(x) for x in [1, 0, 1, 0, 1, 0, 1, 0]])
blacks
```

## Stacked bar charts

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
        # "axes.spines.top": False,
        # "axes.spines.right": False,
        # "axes.spines.left": False,
        # "axes.spines.bottom": False,
    }
)
```

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="lightgrey",
    layout="constrained",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Plot stacked bars
for i in range(0, 8, 2):
    ax.bar(
        x=list("ABCDEFGH"),
        height=blacks,
        color="black",
        width=1,
        bottom=np.ones(len(blacks))*i
    )
    ax.bar(
        x=list("ABCDEFGH"),
        height=~blacks,
        color="black",
        width=1,
        bottom=np.ones(len(blacks))*(i+1)
    )

ax.set_xlim(-0.5, 7.5)
ax.set_ylim(0, 8)
ax.set_yticks(
    ticks=np.arange(0.5, 8, 1),
    labels=[f"{y}" for y in range(1, 9)]
)

# Footer
ax.text(
    1, -0.13,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Scatter plot

```python
xv_odd, yv_odd = np.meshgrid(range(1, 8, 2), range(1, 8, 2))
xv_even, yv_even = np.meshgrid(range(2, 9, 2), range(2, 9, 2))
```

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="lightgrey",
    layout="constrained",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Plot scatter plot

kwargs = {
    "marker": "s",
    "s": 3200,  # marker size
    "c": "black"  # marker colour
}

ax.scatter(
    xv_odd, yv_odd, **kwargs
)
ax.scatter(
    xv_even, yv_even, **kwargs
)

ax.set_xlim(0.5, 8.5)
ax.set_ylim(0.5, 8.5)
ax.set_xticks(
    ticks=np.arange(1, 9, 1),
    labels=list("ABCDEFGH")
)
ax.set_yticks(
    ticks=np.arange(1, 9, 1),
    labels=[f"{y}" for y in range(1, 9)]
)

# Footer
ax.text(
    1, -0.13,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Subplots

```python
fig, axs = plt.subplots(
    8, 8,
    figsize=(8, 8),
    facecolor="lightgrey",
    # sharex=True,
    # sharey=True,
    # layout="constrained",
    subplot_kw={
        "aspect": "equal"
    }
)

fig.subplots_adjust(
    hspace=0,
    wspace=0
)

fig.suptitle(
    "20th of July - International Chess Day",
    fontsize=45,
    weight="bold"
)

# Colour black alternate subplots
for x in range(0, 8, 2):
    for y in range(1, 9, 2):
            axs[x, y].set_facecolor("black")
for x in range(1, 9, 2):
    for y in range(0, 8, 2):
            axs[x, y].set_facecolor("black")

# Remove all ticks from all plots
for a in axs.flat:
    a.set_xticks([])
    a.set_yticks([])

# Add ticks and labels to the bottom row
for a, t in zip(axs[-1, :], list("ABCDEFGH")):
    a.set_xticks(
            ticks=[0.5],
            labels=t
    )

# Add ticks and labels to the first column
for a, t in zip(axs[:, 0], range(8,0, -1)):
    a.set_yticks(
            ticks=[0.5],
            labels=f"{t}"
    )

# Footer
axs[-1, -1].text(
    1, -1.25,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = axs[-1, -1].transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
