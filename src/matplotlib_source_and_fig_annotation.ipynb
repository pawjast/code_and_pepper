## Data

```python
nr_of_samples = 600

x = np.random.rand(nr_of_samples)
y = np.random.rand(nr_of_samples)
size = np.random.randint(1, 150, size=x.shape)
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
        "axes.spines.top": False,
        "axes.spines.right": False
    }
)
```

## Add source and figure description

```python
fig, ax = plt.subplots(
    figsize=(18,9),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "The Dots in `prism` colour scheme",
    fontsize=40,
    weight="bold"
)

ax.scatter(
    x, y,
    s=size,
    c=size,
    cmap="prism"
)

# Footer - figure description
ax.text(
    0.5, -0.12,
    "Fig 1: Some dots",
    transform = ax.transAxes,
    horizontalalignment="center",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=26,
    family="monospace",
    weight="medium"
)

# Footer - source
ax.text(
    0, -0.14,
    "Source: Numpy `rand` function",
    transform = ax.transAxes,
    horizontalalignment="left",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=20,
    family="monospace",
    weight="medium"
)


# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
