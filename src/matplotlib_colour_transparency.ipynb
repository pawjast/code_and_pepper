## Data

```python
n = 200  # number of samples
x = np.linspace(-1.25, 1.25, n) + np.random.normal(loc=0, scale=0.025, size=n)
y1 = x**3 + np.random.normal(loc=0, scale=0.2, size=n) + 0.35
y2 = -x**3 + np.random.normal(loc=0, scale=0.2, size=n) - 0.35
y3 = np.random.normal(loc=0, scale=0.2, size=n)
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
        "axes.spines.top": False,
        "axes.spines.right": False,
        "lines.markersize": 40,
        "lines.markeredgewidth" : 2
    }
)
```

```python
fig, ax = plt.subplots(
    figsize=(18,12),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "Applying `alpha` to colour",
    fontsize=40,
    weight="bold"
)

# Plots
ax.plot(
    x, y1, ".",
    color="blue",
    alpha=0.3
)
ax.plot(
    x, y2, ".",
    color=("red", 0.3)
)
ax.plot(
    x, y3, ".",
    color=(0, 1, 0, 0.3)
)

# Footer
ax.text(
    1, -0.08,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=22,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
