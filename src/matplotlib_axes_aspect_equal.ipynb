## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "xtick.bottom": False,
        "ytick.left": False,
        "grid.color": "#f6f5f4",
        "grid.linewidth": 1.5
    }
)
```

```python
colour_1 = "#f5c211"
colour_2 = "#c061cb"
```

```python
fig, axs = plt.subplots(
    1, 3,
    figsize=(20, 12),
    facecolor="white",
    layout="constrained"
)

axs[0].set_xlim(0, 1)
axs[0].set_ylim(0, 1)

# `aspect` == auto
circ = mpl.patches.Circle(
    (0.7, 0.7), 0.2,
    color=colour_1
)
rect = mpl.patches.Rectangle(
    (0.1, 0.1), 0.4, 0.4,
    color=colour_2
)
axs[0].add_artist(circ)
axs[0].add_artist(rect)

# `aspect` == 'equal'
axs[1].set_aspect("equal")
circ1 = mpl.patches.Circle(
    (0.3, 0.7), 0.2,
    color=colour_1
)
rect1 = mpl.patches.Rectangle(
    (0.5, 0.1), 0.4, 0.4,
    color=colour_2
)
axs[1].add_artist(circ1)
axs[1].add_artist(rect1)

# `aspect` == number (ratio of x to y)
axs[2].set_aspect(0.5)  # 1x = 0.2y
circ2 = mpl.patches.Circle(
    (0.3, 0.7), 0.2,
    color=colour_1
)
rect2 = mpl.patches.Rectangle(
    (0.5, 0.1), 0.4, 0.4,
    color=colour_2
)
axs[2].add_artist(circ2)
axs[2].add_artist(rect2)

for ax in axs:
    ax.set_axisbelow(True)
    ax.set_xticks(np.arange(0, 1.1, 0.2)[1:-1])
    ax.set_xticks(np.arange(0.1, 1.1, 0.2), minor=True)
    ax.set_yticks(np.arange(0, 1.1, 0.2)[1:-1])
    ax.set_yticks(np.arange(0.1, 1.1, 0.2), minor=True)
    ax.grid(which="both")

axs[0].set_title("aspect == `auto`")
axs[1].set_title("aspect == `equal`")
axs[2].set_title("aspect == `<float>`")

# Footer
axs[2].text(
    1, -1.55,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = axs[2].transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
