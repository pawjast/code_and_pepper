## Data

```python

x = np.random.rand(25, 25)
y = np.random.rand(25, 25)
```

## Plot

### Settings

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
        "axes.spines.top": False,
        "axes.spines.right": False,
        # "axes.spines.left": False,
        # "axes.spines.bottom": False,
    }
)
```

```python
my_marker_1 = r"\clubsuit"  # from mathtext
my_marker_2 = u"\u266A"  # from unicode, musical note
my_marker_3 = r"☘️"  # emoji - doesn't work with all emojis - OS dependent??
```

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "St Patrick's Day",
    fontsize=40,
    weight="bold"
)

ax.axis("Off")

# YOUR PLOT HERE

ax.scatter(
    x, y,
    marker=f"${my_marker_3}$",
    c=("green", 0.5),
    s=500
)
ax.scatter(
    x+1, y,
    marker=f"${my_marker_2}$",
    c=([0.9, 0.9, 0.9], 0.5),
    s=500
)
ax.scatter(
    x+2, y,
    marker=f"${my_marker_1}$",
    c=("orange", 0.5),
    s=500
)

# Footer
ax.text(
    1, 0,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
