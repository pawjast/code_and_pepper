## Data

```python
data = {
    "a": 10,
    "b": 12,
    "c": 7,
    "d": 18,
    "e": 3
}

im = plt.imread("meme_logo.png")
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
        "axes.spines.top": False,
        "axes.spines.right": False,
        # "axes.spines.left": False,
        # "axes.spines.bottom": False,
    }
)
```

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "Image watermarked graph",
    fontsize=40,
    weight="bold"
)

ax.bar(
    x=data.keys(),
    height=data.values(),
    color="green"
)

fig.figimage(
    im,
    alpha=0.2,
    resize=True
)

# Footer
ax.text(
    1, -0.28,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
