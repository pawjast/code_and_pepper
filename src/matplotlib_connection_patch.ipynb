# `matplotlib`: example of using `ConnectionPatch` object

https://matplotlib.org/stable/api/_as_gen/matplotlib.patches.ConnectionPatch.html#matplotlib.patches.ConnectionPatch

It's capable of connecting two points located in two different `Axes`.

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import ConnectionPatch
```

## Settings

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
    }
)

marker_size_val = 15
link_thickness = 2
```

## Data

```python
point_a1 = (0.7, 0.7)
point_a2 = (0.5, 0.7)
point_b1 = (0.5, 0.7)
point_b2 = (0.7, 0.7)
point_c1 = (0.3, 0.5)
point_c2 = (0.3, 0.3)
```

## Plot 1 A

`ConnectionPatch` object gets added to the `Axes` using `ax.add_artist()` method.

```python
mosaic = """
    ab
"""

fig, axs = plt.subplot_mosaic(
    mosaic=mosaic,
    figsize=(12, 8),
    facecolor="white",
    sharex=True,
    sharey=True,
)

# Titles
fig.suptitle(
    "Using `ConnectionPatch` to link points in different Axes",
    fontsize=25,
    weight="bold"
)
axs["a"].set_title("a")
axs["b"].set_title("b")

axs["a"].plot(
    *point_a1,
    marker="s",
    markersize=marker_size_val,
    color="blue"
)
axs["b"].plot(
    *point_b1,
    marker="o",
    markersize=marker_size_val,
    color="green"
)


# Settings
axs["a"].set_xlim(0, 1)
axs["a"].set_ylim(0, 1)

for ax_name in axs:
    axs[ax_name].set_xticks(np.arange(0.2, 1, 0.2))
    axs[ax_name].set_yticks(np.arange(0.2, 1, 0.2))
    axs[ax_name].grid()

# Links between points
connection1 = ConnectionPatch(
    xyA=point_a1,
    coordsA=axs["a"].transData,
    xyB=point_b1,
    coordsB=axs["b"].transData,
    linewidth=link_thickness,
    color="red"
)
axs["b"].add_artist(connection1)

# Footer
axs["b"].text(
    1, -0.15,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = axs["b"].transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Plot 2

The 2x2 grid layout example with some challenges when adding artist to the `Axes`.

```python
mosaic = """
    ab
    cd
"""

fig, axs = plt.subplot_mosaic(
    mosaic=mosaic,
    figsize=(12, 12),
    facecolor="white",
    sharex=True,
    sharey=True,
)

# Titles
fig.suptitle(
    "Using `ConnectionPatch` to link points in different Axes",
    fontsize=25,
    weight="bold"
)
axs["a"].set_title("a")
axs["b"].set_title("b")
axs["c"].set_title("c")
axs["d"].set_title("d")

axs["a"].plot(
    *point_a1,
    marker="s",
    markersize=marker_size_val,
    color="blue"
)
axs["a"].plot(
    *point_a2,
    marker="s",
    markersize=marker_size_val,
    color="blue"
)
axs["b"].plot(
    *point_b1,
    marker="o",
    markersize=marker_size_val,
    color="green"
)
axs["b"].plot(
    *point_b2,
    marker="o",
    markersize=marker_size_val,
    color="green"
)
axs["c"].plot(
    *point_c1,
    marker="^",
    markersize=marker_size_val,
    color="orange"
)
axs["c"].plot(
    *point_c2,
    marker="^",
    markersize=marker_size_val,
    color="orange"
)

# Settings
axs["a"].set_xlim(0, 1)
axs["a"].set_ylim(0, 1)

for ax_name in axs:
    axs[ax_name].set_xticks(np.arange(0.2, 1, 0.2))
    axs[ax_name].set_yticks(np.arange(0.2, 1, 0.2))
    axs[ax_name].grid()

# Links between points
connection1 = ConnectionPatch(
    xyA=point_a1,
    coordsA=axs["a"].transData,
    xyB=point_b1,
    coordsB=axs["b"].transData,
    linewidth=link_thickness,
    color="red"
)
axs["b"].add_artist(connection1)

connection2 = ConnectionPatch(
    xyA=point_b2,
    coordsA=axs["b"].transData,
    xyB=point_c2,
    coordsB=axs["c"].transData,
    linewidth=link_thickness,
    color="red",
    connectionstyle="angle3"

)
axs["c"].add_artist(connection2)

connection3 = ConnectionPatch(
    xyA=point_c1,
    coordsA=axs["c"].transData,
    xyB=point_a2,
    coordsB=axs["a"].transData,
    linewidth=link_thickness,
    color="red",
    connectionstyle="angle"

)
axs["a"].add_artist(connection3)


# Footer
axs["d"].text(
    1, -0.2,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = axs["d"].transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Plot 3

The 2x2 grid layout example with some when the artist is added to the `Figure`.  
Also, using `fig.add_artist()` (as opposed to `ax.add_artist()`) allows to use `layout="constrained"` when creating `Figure`.

```python
mosaic = """
    ab
    cd
"""

fig, axs = plt.subplot_mosaic(
    mosaic=mosaic,
    figsize=(12, 12),
    facecolor="white",
    sharex=True,
    sharey=True,
    layout="constrained"
)

# Titles
fig.suptitle(
    "Using `ConnectionPatch` to link points in different Axes",
    fontsize=25,
    weight="bold"
)
axs["a"].set_title("a")
axs["b"].set_title("b")
axs["c"].set_title("c")
axs["d"].set_title("d")

axs["a"].plot(
    *point_a1,
    marker="s",
    markersize=marker_size_val,
    color="blue"
)
axs["a"].plot(
    *point_a2,
    marker="s",
    markersize=marker_size_val,
    color="blue"
)
axs["b"].plot(
    *point_b1,
    marker="o",
    markersize=marker_size_val,
    color="green"
)
axs["b"].plot(
    *point_b2,
    marker="o",
    markersize=marker_size_val,
    color="green"
)
axs["c"].plot(
    *point_c1,
    marker="^",
    markersize=marker_size_val,
    color="orange"
)
axs["c"].plot(
    *point_c2,
    marker="^",
    markersize=marker_size_val,
    color="orange"
)

# Settings
axs["a"].set_xlim(0, 1)
axs["a"].set_ylim(0, 1)

for ax_name in axs:
    axs[ax_name].set_xticks(np.arange(0.2, 1, 0.2))
    axs[ax_name].set_yticks(np.arange(0.2, 1, 0.2))
    axs[ax_name].grid()

# Links between points
connection1 = ConnectionPatch(
    xyA=point_a1,
    coordsA=axs["a"].transData,
    xyB=point_b1,
    coordsB=axs["b"].transData,
    linewidth=link_thickness,
    color="red"
)
fig.add_artist(connection1)

connection2 = ConnectionPatch(
    xyA=point_b2,
    coordsA=axs["b"].transData,
    xyB=point_c2,
    coordsB=axs["c"].transData,
    linewidth=link_thickness,
    color="red",
    connectionstyle="angle3"

)
fig.add_artist(connection2)

connection3 = ConnectionPatch(
    xyA=point_c1,
    coordsA=axs["c"].transData,
    xyB=point_a2,
    coordsB=axs["a"].transData,
    linewidth=link_thickness,
    color="red",
    connectionstyle="angle"

)
fig.add_artist(connection3)


# Footer
axs["d"].text(
    1, -0.2,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = axs["d"].transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
