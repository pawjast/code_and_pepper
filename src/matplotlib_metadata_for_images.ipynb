# `matplotlib` metadata for images

## Imports

```python
import matplotlib.pyplot as plt
```

## Create dictionary of metadata

```python
from datetime import date

metadata_dict = {
    "Title": "Bar chart with metadata in it",
    "Author": "Pawel Jastrzebski (pawjast)",
    "Description": "Simple bar chart created in Matplotlib",
    "Creation Time": f"{date.today():%d %B %Y}",
    "Source": "https://twitter.com/pawjast",
    "Copyright": "All rights reserved.",
    "Disclaimer": "Nothing to worry about",
    "Warning": "This is harmless bar chart.",
    "Comments": "...",
    "my_own_tag": "Hi!"
}
```

## Plot

```python
fig, ax = plt.subplots(
    figsize=(16,10),
    facecolor="white"
)

# basic bar chart
ax.bar(
    x = list("abcde"),
    height=[1, 6, 4, 2, 3],
    color="#57e389"
)

ax.plot(
    [2, 5, 3, 4, 2],
    marker="o",
    markersize=12,
    linewidth=2,
    color="#e01b24"
)

# Remove spines
for position in ["left", "top", "right"]:
    ax.spines[position].set_visible(False)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="y",
    which="major",
    left=False,
    labelleft=False
)

# Update apperance of the `x` axis tick labels
ax.tick_params(
    axis="x",
    labelsize=20
)

plt.tight_layout()

plt.savefig(
    fname="plot_with_metadata.png",
    format="png",
    dpi=125,
    metadata=metadata_dict
);
```
