```python
import matplotlib.pyplot as plt
import numpy as np
```

## Data

```python
x_blue = np.random.rand(400)
y_blue = np.random.rand(400)
x_red = np.random.rand(800)
y_red = np.random.rand(800)
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 18,
        "axes.spines.right": False,
        "axes.spines.top": False,
    }
)
```

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "THE DOTS",
    fontsize=30,
    weight="bold",
    x=0,
    y=1.075,
    ha="left",
    color="#3d3846",
    transform = ax.transAxes
)

ax.set_title(
    "There are more red dots than the blue ones",
    loc="left",
    color="#77767b",
)

ax.scatter(
    x=x_blue,
    y=y_blue,
    s=10,
    label="Blues",
    marker="o",
    edgecolor="blue",
    alpha=0.7
)
ax.scatter(
    x=x_red,
    y=y_red,
    s=10,
    label="Reds",
    marker="o",
    edgecolor="red",
    alpha=0.7
)


# Footer
ax.text(
    1, -0.09,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="1.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict,
#     bbox_inches="tight"
# );
```

## Plot 2
With colour accents in the text

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "THE DOTS",
    fontsize=30,
    weight="bold",
    x=0,
    y=1.075,
    ha="left",
    color="#3d3846",
    transform = ax.transAxes
)

# Use `text` as a suptitle
txt = ax.text(
    0, 1.01,
    "There are more",
    color="#77767b",
    transform = ax.transAxes
)
txt = ax.annotate(
    " red",
    xycoords=txt,
    xy=(1, 0),
    verticalalignment="bottom",
    color="red"
)
txt = ax.annotate(
    " dots than the",
    xycoords=txt,
    xy=(1, 0),
    verticalalignment="bottom",
    color="#77767b"
)
txt = ax.annotate(
    " blue",
    xycoords=txt,
    xy=(1, 0),
    verticalalignment="bottom",
    color="blue"
)
txt = ax.annotate(
    " ones",
    xycoords=txt,
    xy=(1, 0),
    verticalalignment="bottom",
    color="#77767b",
)

ax.scatter(
    x=x_blue,
    y=y_blue,
    s=10,
    label="Blues",
    marker="o",
    edgecolor="blue",
    alpha=0.7
)
ax.scatter(
    x=x_red,
    y=y_red,
    s=10,
    label="Reds",
    marker="o",
    edgecolor="red",
    alpha=0.7
)


# Footer
ax.text(
    1, -0.09,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="2.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict,
#     bbox_inches="tight"
# );
```
