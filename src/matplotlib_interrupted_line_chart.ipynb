# Interrupted line plot in `matplotlib`

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [0.5, 1.5, None, 1, 1.75, 2]
```

## Plot the data

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor="#f6f5f4"
)
fig.suptitle(
    "Interrupted plot line: plotted from lists",
    fontsize=16
)

ax.plot(
    x, y,
    color="black",
    marker="o"
)

plt.tight_layout()

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```

## Plot the data from `DataFrame`

This is only to check that the data stored in a `DataFrame` will still plot as before.

```python
df = pd.DataFrame(
    y,
    columns=["y"],
    index=x
)

df
```

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor="#f6f5f4"
)
fig.suptitle(
    "Interrupted plot line: plotted from df",
    fontsize=16
)

ax.plot(
    df,
    color="black",
    marker="o"
)

plt.tight_layout()
plt.savefig(
    fname="plot2.png",
    format="png",
    dpi=125
);
```

## Drop lines

I used to use this "feature" to create drop lines in Excel.

### Extra import

```python
import random
```

### Data for the line plot

```python
nr_of_items = 20
x = list(range(nr_of_items))
random.seed(0)
y = random.choices(range(2, 8), k=nr_of_items)
```

### Data for the drop lines

```python
# repeat each `x` 3 times
x_drop = [x_val for x_val in x for _ in range(3)]
# extend each `y` with 0 and None
y_drop = [y1 for y_val in y for y1 in [y_val, 0, None]]
```

### Plot the drop lines only

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor="#f6f5f4"
)
fig.suptitle(
    "The drop lines only",
    fontsize=16
)

ax.plot(
    x_drop, y_drop,
    color="black",
    linewidth=.5,
    linestyle="--"
)
ax.set_ylim(0, 8)
ax.set_xticks(x)


plt.tight_layout()
plt.savefig(
    fname="plot3.png",
    format="png",
    dpi=125
);
```

### Plot the line plot and the drop lines

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor="#f6f5f4"
)
fig.suptitle(
    "Line plot with the drop lines",
    fontsize=16
)

ax.plot(
    x, y,
    color="black",
    marker="o"
)
ax.plot(
    x_drop, y_drop,
    color="black",
    linewidth=.5,
    linestyle="--"
)
ax.set_ylim(0, 8)
ax.set_xticks(x)

plt.tight_layout()
plt.savefig(
    fname="plot4.png",
    format="png",
    dpi=125
);
```

### Plot the line plot and the drop lines 2

Skip the line where the same consecutive `y`.

```python
# PLOT LINE
nr_of_items = 20
x = list(range(nr_of_items))
random.seed(4)
y = random.choices(range(2, 5), k=nr_of_items)

# Remove middle consecutive `y` values
y_with_none = []
for i, val in enumerate(y[1:-1], start=1):
    if y[i-1] == val == y[i+1]:
        y_with_none.append(None)
    else:
        y_with_none.append(val)

y_with_none.insert(0, y[0])
y_with_none.append(y[-1])

# DROP LINES
# repeat each `x` 3 times
x_drop = [x_val for x_val in x for i in range(3)]
# extend each `y` with 0 and None
y_drop = [y1 for y_val in y_with_none for y1 in [y_val, 0, None]]
```

```python
fig, ax = plt.subplots(
    figsize=(9, 6),
    facecolor="#f6f5f4"
)
fig.suptitle(
    "Line plot with the drop lines (for changing values only)",
    fontsize=16
)

ax.plot(
    x, y,
    color="black",
    marker="o"
)
ax.plot(
    x_drop, y_drop,
    color="black",
    linewidth=.5,
    linestyle="--"
)
ax.set_ylim(0, 5)
ax.set_xticks(x)

plt.tight_layout()
plt.savefig(
    fname="plot5.png",
    format="png",
    dpi=125
);
```
