## Data

```python
nr_of_samples = 450

x = np.random.rand(nr_of_samples)
y = np.random.rand(nr_of_samples)
size = np.random.randint(1, 150, size=x.shape)
```

## Using `rcParams` - individually

```python
rcParams["figure.facecolor"] = "black"
rcParams["text.color"] = "white"
rcParams["xtick.color"] = "white"
rcParams["ytick.color"] = "white"
rcParams["font.size"] = 14
rcParams["axes.prop_cycle"] = cycler(color=["black"])
rcParams["axes.facecolor"] = "#e01b24"
rcParams["grid.linewidth"] = 0.25
rcParams["grid.color"] = "white"
```

```python
fig, ax = plt.subplots(
    figsize=(8, 6),
    layout="constrained"
)
fig.suptitle(
    "Making Seedy Watermelon Pattern",
    fontsize=20,
    fontfamily="serif",
    weight="bold"
)

ax.scatter(
    x, y,
    s=size
)
ax.grid()

# Footer
ax.text(
    1, -0.13,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    fontsize=16,
    family="monospace",
    weight="medium"
);
```

```python
# reset settings
rcdefaults()
```

## Using context manager

```python
with plt.rc_context(
    rc={
        "figure.facecolor": "black",
        "text.color": "white",
        "xtick.color": "white",
        "ytick.color": "white",
        "font.size": 14,
        "axes.prop_cycle": cycler(color=["black"]),
        "axes.facecolor": "#e01b24",
        "grid.linewidth": 0.25,
        "grid.color": "white"
    }
):
    fig, ax = plt.subplots(
        figsize=(8, 6),
        layout="constrained"
    )
    fig.suptitle(
        "Making Seedy Watermelon Pattern",
        fontsize=20,
        fontfamily="serif",
        weight="bold"
    )

    ax.scatter(
        x, y,
        s=size
    )
    ax.grid()

    # Footer
    ax.text(
        1, -0.13,
        "by @pawjast\nmade with @matplotlib \u2665",
        transform = ax.transAxes,
        horizontalalignment="right",
        verticalalignment="bottom",
        fontsize=16,
        family="monospace",
        weight="medium"
    );
```

```python
# confirm default settings are applied once out of the context manager
fig, ax = plt.subplots(
    figsize=(8, 6),
    layout="constrained"
)
fig.suptitle(
    "Making Seedy Watermelon Pattern",
    fontsize=20,
    fontfamily="serif",
    weight="bold"
)

ax.scatter(
    x, y,
    s=size
)
ax.grid()

# Footer
ax.text(
    1, -0.13,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);
```
