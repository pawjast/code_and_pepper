## Settings

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "lines.linewidth": 2,
        "grid.linewidth": 1,
        "grid.color": "#f6f5f4",
        "xtick.bottom": False,
        "ytick.left": False,
    }
)
```

```python
colour_1 = "#33d17a"
colour_2 = "#3584e4"
colour_3 = "#e01b24"
colour_4 = "#9141ac"
colour_5 = "#ff7800"
```

## Plot

```python
mosaic = """
   ab
   cd
"""

fig, axs = plt.subplot_mosaic(
    mosaic=mosaic,
    figsize=(16,12),
    facecolor="white",
    sharex=True,
    sharey=True,
    layout="constrained"
)
# fig.suptitle(
#     "TITLE",
#     fontsize=25,
#     weight="bold"
# )

# Settings
axs["a"].set_xlim(0, 1)
axs["a"].set_ylim(0, 1)

for ax_name in axs:
    axs[ax_name].set_xticks(np.linspace(0, 1, 6)[1:-1])
    axs[ax_name].set_yticks(np.linspace(0, 1, 6)[1:-1])
    axs[ax_name].grid()
    axs[ax_name].set_axisbelow(True)


# PLOTS
# Horizontal lines
axs["a"].axhline(
    y=0.2,
    color=colour_1
)
axs["a"].axhline(
    y=0.4,
    xmax=0.3,
    color=colour_1
)
axs["a"].axhline(
    y=0.6,
    xmin=0.3,
    xmax=0.6,
    color=colour_1
)
axs["a"].axhline(
    y=0.8,
    xmin=0.6,
    color=colour_1
)

# Vertical lines
axs["b"].axvline(
    x=0.2,
    color=colour_2
)
axs["b"].axvline(
    x=0.4,
    ymax=0.3,
    color=colour_2
)
axs["b"].axvline(
    x=0.6,
    ymin=0.3,
    ymax=0.6,
    color=colour_2
)
axs["b"].axvline(
    x=0.8,
    ymin=0.6,
    color=colour_2
)

# Ax spans
axs["c"].axhspan(
    ymin=0.1,
    ymax=0.2,
    color=colour_3
)
axs["c"].axhspan(
    ymin=0.6,
    ymax=0.65,
    xmin=0.1,
    xmax=0.5,
    color=colour_3
)
axs["c"].axvspan(
    xmin=0.75,
    xmax=0.95,
    color=colour_4
)
axs["c"].axvspan(
    xmin=0.6,
    xmax=0.7,
    ymin=0.4,
    color=colour_4
)

# Angled lines
axs["d"].axline(
    xy1=(0, 0),
    xy2=(1, 1),
    color=colour_5
)
axs["d"].axline(
    xy1=(0.5, .50), 
    slope=-1,
    color=colour_5
)

# Footer
axs["d"].text(
    1, -.16,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = axs["d"].transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
