# `matplotlib` creating inset plot

## Imports

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
import numpy as np
```

## Data

```python
nr_of_samples = 500

x = np.linspace(0, 4, nr_of_samples)
y1 = np.power(x, 2)*np.sin(x)
y2 = -2*np.power(x, 2)*np.sin(x*0.5) + 5

y1_scat = y1+np.random.randn(len(y1))
y2_scat = y2+np.random.randn(len(y2))
```

## Plot

```python
# inset limits
xlim_vals = (1, 2)
ylim_vals = (-3, 6)

# inset colour
inset_col = (1, 1, 0, 0.2)
```

```python
fig, ax = plt.subplots(
    figsize=(16, 12),
    facecolor="white",
    layout="constrained"
)
fig.suptitle(
    "SOME PLOT",
    fontsize=30,
    fontfamily="serif",
    weight="bold"
)

rect1 = 0.08, 0.16, 0.4, 0.4 #  x, y, width, heigh
ax1 = fig.add_axes(
    rect1, 
    facecolor=inset_col,
)

for a in [ax, ax1]:
    a.plot(
        x, y1,
        lw=2,
        color="#ff7800",
        label="line 1"
    )
    a.plot(
        x, y2,
        lw=2,
        color="#c061cb",
        label="line 2"
    )

    a.scatter(
        x, y1_scat,
        marker="o",
        edgecolors="#ff7800",
        facecolor="white"
    )
    a.scatter(
        x, y2_scat,
        marker="o",
        edgecolors="#c061cb",
        facecolor="white"
    )

poly = Polygon(
    [
        [xlim_vals[0], ylim_vals[0]], [xlim_vals[1], ylim_vals[0]], [xlim_vals[1], ylim_vals[1]], [xlim_vals[0], ylim_vals[1]]
    ],
    closed=True,
    color=inset_col
)
ax.add_patch(poly)

ax1.set_title(
    "Inset Plot",
    fontsize=20
)
ax1.set_xlim(xlim_vals)
ax1.set_ylim(ylim_vals)

ax.legend(
    fontsize=20,
    frameon=False
)

# Footer
ax.text(
    1, -0.09,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="pawjast.png",
#     format="png",
#     dpi=125
# );
```
