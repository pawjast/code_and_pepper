# `matplotlib` using `ax.set()`

## Imports

```python
import matplotlib.pyplot as plt
```

## Plot 1

Set all properies separately.

```python
fig, ax = plt.subplots(
    figsize=(9, 9),
    facecolor="white"
)

ax.set_aspect("equal")
ax.set_facecolor("#f6f5f4")
ax.set_title("Using `ax.set_<property name>`")
ax.set_xlabel("The X's")
ax.set_ylabel("The Y's")
ax.set_xticks([0, 1, 2, 3])
ax.set_yticks([0, 1, 2, 3])
ax.set_xticklabels(list("abcd"))

plt.tight_layout()

# plt.savefig(
#     fname="1.png",
#     format="png",
#     dpi=125
# );
```

```python
fig, ax = plt.subplots(
    figsize=(9, 9),
    facecolor="white"
)

ax.set(
    aspect="equal",
    facecolor="#f6f5f4",
    title="Using `ax.set()`",
    xlabel="The X's",
    ylabel="The Y's",
    xticks=[0, 1, 2, 3],
    yticks=[0, 1, 2, 3],
    xticklabels=list("abcd")
)

plt.tight_layout()

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125
# );
```
