# `matplotlib` 3D plots

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Create 3D axis

### Way 1: figure + add_subplot

```python
fig = plt.figure(
    figsize=(10, 10),
    facecolor="#f6f5f4",
    layout="constrained"
)

fig.suptitle(
    "projection='3d'",
    fontsize=24,
)

ax = fig.add_subplot(
    projection="3d"
)

plt.savefig(
    fname="1.png",
    format="png",
    dpi=125
);
```

### Way 2: subplots

```python
fig, ax = plt.subplots(
    figsize=(16, 12),
    subplot_kw={
        "projection": "3d"
    }
)
```

### Way 3: plt.subplot_mosaic

```python
fig, ax = plt.subplot_mosaic(
    [
        ["3d"]
    ],
    figsize=(16, 12),
    subplot_kw={
        "projection": "3d"
    }
)
```

### Way 4: fig.subplot_mosaic

```python
fig = plt.figure(
    figsize=(16, 12)
)

ax = fig.subplot_mosaic(
    """
    A
    """,
    subplot_kw={
        "projection": "3d"
    }
)
```

## Line plot

### Data

```python
step = 0.05
z1 = np.arange(0, 75, step)
r1 = np.linspace(1, 0.01, len(z1))
x1 = r1**2 * np.sin(z1)
y1 = r1**2 * np.cos(z1)

z2 = np.arange(75, 150, step)
r2 = np.linspace(1, 0.01, len(z2))
x2 = r1[::-1]**2 * np.sin(z2)
y2 = r1[::-1]**2 * np.cos(z2)
```

### Plot

```python
fig = plt.figure(
    figsize=(10, 10),
    facecolor="#f6f5f4",
    layout="constrained"
)

ax = fig.add_subplot(
    projection="3d"
)
ax.set_aspect("equal")

fig.suptitle(
    "Line Plot",
    fontsize=24,
)

ax.plot(
    x1, y1, z1,
    color="#c061cb",  # purple
    linewidth=1.5
)

ax.plot(
    x2, y2, z2,
    color="#9141ac",  # purple
    linewidth=1.5
)

ax.view_init(
    elev=15,
    azim=15,
    roll=0
    )

for a in (ax.xaxis, ax.yaxis, ax.zaxis):
    a.pane.set_color("#f6f5f4")
    a.pane.set_edgecolor("#9a9996")
    a.line.set_color((1, 1, 1, 0))

# Remove ticks and tick labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])


ax.grid(False)

plt.savefig(
    fname="2.png",
    format="png",
    dpi=125
);
```

## Scatter plot

### Data

```python
size = 4000

x = np.random.uniform(
    low=-1,
    high=1,
    size=size
)
y = np.random.uniform(
    low=-1,
    high=1,
    size=size
)
z = np.random.uniform(
    low=-1,
    high=1,
    size=size
)

r = 0.75

flag = ((x**2 + y**2 + z**2)**(-2) > r )
```

### Plot

```python
fig = plt.figure(
    figsize=(10, 10),
    facecolor="#f6f5f4",
    layout="constrained"
)


ax = fig.add_subplot(
    projection="3d"
)
ax.set_aspect("equal")

fig.suptitle(
    "Scatter Plot",
    fontsize=24,
)

ax.scatter(
    x[flag], y[flag], z[flag],
    color="#c01c28",  # red
)
ax.scatter(
    x[np.invert(flag)], y[np.invert(flag)], z[np.invert(flag)],
    color="#3584e4",  # blue
)

ax.view_init(
    elev=15,
    azim=15,
    roll=0
    )

for a in (ax.xaxis, ax.yaxis, ax.zaxis):
    a.pane.set_color("#f6f5f4")
    a.pane.set_edgecolor("#9a9996")
    a.line.set_color((1, 1, 1, 0))

# Remove ticks and tick labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])


ax.grid(False)

plt.savefig(
    fname="3.png",
    format="png",
    dpi=125
);
```

## Surface plot

### Data

```python
size = 75
x = np.linspace(-3, 3, size)
y = np.linspace(-3, 3, size)
X, Y = np.meshgrid(x, y)
Z1 = (X**2 + Y**2) -18
Z2 = -(X**2 + Y**2) + 18
```

### Plot

```python
fig = plt.figure(
    figsize=(10, 10),
    facecolor="#f6f5f4",
    layout="constrained"
)


ax = fig.add_subplot(
    projection="3d"
)
ax.set_aspect("equal")

fig.suptitle(
    "Surface Plot",
    fontsize=24,
)

ax.plot_surface(
    X, Y, Z1,
    color="#ff7800",  # orange
    alpha=0.65,
    edgecolor="#dc8add"
)
ax.plot_surface(
    X, Y, Z2,
    color="#f5c211",  # yellow
    alpha=0.65,
    edgecolor="#dc8add"
)

ax.view_init(
    elev=15,
    azim=15,
    roll=0
    )

for a in (ax.xaxis, ax.yaxis, ax.zaxis):
    a.pane.set_color("#f6f5f4")
    a.pane.set_edgecolor("#9a9996")
    a.line.set_color((1, 1, 1, 0))

# Remove ticks and tick labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])


ax.grid(False)

plt.savefig(
    fname="4.png",
    format="png",
    dpi=125
);
```

## Wireframe

### Data

```python
size = 25
x = np.linspace(0.25, 1, size)
y = np.linspace(-1, 1, size)
X, Y = np.meshgrid(x, y)
Z = Y ** 2 / X ** 2
```

### Plot

```python
fig = plt.figure(
    figsize=(10, 10),
    facecolor="#f6f5f4",
    layout="constrained"
)

ax = fig.add_subplot(
    projection="3d"
)
ax.set_aspect("equal")

fig.suptitle(
    "Wireframe Plot",
    fontsize=24,
)

ax.plot_wireframe(
    X, Y, Z,
    color="#26a269",  # green
)

ax.view_init(
    elev=15,
    azim=15,
    roll=0
    )

for a in (ax.xaxis, ax.yaxis, ax.zaxis):
    a.pane.set_color("#f6f5f4")
    a.pane.set_edgecolor("#9a9996")
    a.line.set_color((1, 1, 1, 0))

# Remove ticks and tick labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])


ax.grid(False)

plt.savefig(
    fname="5.png",
    format="png",
    dpi=125
);
```
