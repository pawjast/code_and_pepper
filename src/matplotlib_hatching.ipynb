# Hatch fill in `matplotlib` plots

[matplotlib hatch style reference](https://matplotlib.org/stable/gallery/shapes_and_collections/hatch_style_reference.html)

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## fill_between plot

### Data

```python
step = 0.1
x = np.arange(-25, 25, step)
y = x**3
```

### Plot

```python
line_width_val = 2

fig, ax = plt.subplots(
    figsize=(9,6),
    facecolor="white"
)
fig.suptitle(
    "fill_between with hatching",
    fontsize=20,
)

# Plot line
ax.plot(
    x, y,
    linewidth=line_width_val,
    color="black"
)

# Plot fill
ax.fill_between(
    x, y,
    color="white",
    edgecolor="black",
    linewidth=0,
    hatch="x"
)

# Set axis labels
ax.set_xlabel(
    "x",
    fontsize=16,
    loc="right"
)
ax.set_ylabel(
    "y",
    fontsize=16,
    loc="top",
    rotation="horizontal"
)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="both",
    which="major",
    left=False,
    labelleft=False,
    bottom=False,
    labelbottom=False
)

# Update spines thickness and move them to the center
for position in ["left", "bottom"]:
    ax.spines[position].set_position("center")
    ax.spines[position].set_linewidth(line_width_val)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```

## Scatter plot

### Data

```python
nr_of_items = 10

np.random.seed(2)
df = pd.DataFrame(
    {
        "x": np.random.randint(1, 300, nr_of_items),
        "y": np.random.randint(1, 300, nr_of_items),
        "s": np.random.randint(1, 500, nr_of_items)*10  # size
    }
)

df
```

### Plot

```python
line_width_val = 2

fig, ax = plt.subplots(
    figsize=(9,6),
    facecolor="white"
)
fig.suptitle(
    "scatter plot with hatching",
    fontsize=20,
)

# Plot scatter
ax.scatter(
    df["x"],
    df["y"],
    df["s"],
    color="white",
    edgecolor="black",
    linewidth=line_width_val,
    hatch="*"
)

# Set axis labels
ax.set_xlabel(
    "x",
    fontsize=16
)
ax.set_ylabel(
    "y",
    fontsize=16
)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="both",
    which="major",
    left=False,
    labelleft=False,
    bottom=False,
    labelbottom=False
)

# Update spines thickness
for position in ["left", "top", "right", "bottom"]:
    ax.spines[position].set_linewidth(line_width_val)

plt.tight_layout()

plt.savefig(
    fname="plot2.png",
    format="png",
    dpi=125
);
```

## Bar Chart

### Data

```python
df = pd.DataFrame.from_dict(
    {
        "Q1": [12, 15],
        "Q2": [10, 8],
        "Q3": [11, 11],
        "Q4": [13, 14]
    },
    orient="index",
    columns=["2021", "2022"]
)

df
```

### Plot

```python
x = np.arange(len(df))
width = 0.4
multiplier = 0
line_width_val = 2

fig, ax = plt.subplots(
    figsize=(9,6),
    facecolor="white"
)
fig.suptitle(
    "bar chart with hatching",
    fontsize=20,
)

hatch_patterns = [".", "-"]

# Plot bars
for col, hatch_type in zip(df, hatch_patterns):
    offset = width * multiplier
    ax.bar(
        x=x+offset,
        height=df[col],
        width=width,
        color="white",
        edgecolor="black",
        linewidth=line_width_val,
        hatch=hatch_type
    )
    multiplier += 1

# Set axis ticks and tick labels
ax.set_xticks(
    x+width/2,
    df.index,
    fontsize=14
)

# Set axis limit
ax.set_ylim(0, 16)

# Remove spines
for position in ["left", "top", "right"]:
    ax.spines[position].set_visible(False)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="y",
    which="major",
    left=False,
    labelleft=False
)

# Update `x` axis tick thickness
ax.tick_params(
    axis="x",
    which="major",
    width=line_width_val,
    labelsize=16
)

ax.spines["bottom"].set_linewidth(line_width_val)

plt.tight_layout()

plt.savefig(
    fname="plot3.png",
    format="png",
    dpi=125
);
```

## Hatchinng with colours

### Data

```python
step = 0.1
x = np.arange(-10, 10, step)
y = x**2 + 10
```

### Plot

```python
line_width_val = 2

fig, ax = plt.subplots(
    figsize=(9,6),
    facecolor="white"
)
fig.suptitle(
    "fill_between with hatching and colour background",
    fontsize=20,
)

# Plot line
ax.plot(
    x, y,
    linewidth=line_width_val,
    color="black"
)

# Plot fill
ax.fill_between(
    x, y,
    color="#f9f06b",  # purple 1
    edgecolor="black",
    linewidth=0,
    hatch="+"
)

# Set axis limit
ax.set_xlim(-9, 9)
ax.set_ylim(0, 100)

# Set axis labels
ax.set_xlabel(
    "x",
    fontsize=16,
    loc="right"
)
ax.set_ylabel(
    "y",
    fontsize=16,
    loc="top",
    rotation="horizontal"
)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="both",
    which="major",
    left=False,
    labelleft=False,
    bottom=False,
    labelbottom=False
)

# Update spines thickness
for position in ["left", "bottom"]:
    ax.spines[position].set_linewidth(line_width_val)

# Move "y" spine to the center
ax.spines["left"].set_position("center")

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot4.png",
    format="png",
    dpi=125
);
```

## Dense hatching

### Data

```python
x = range(3)
y1 = [1]*len(x)
y2 = [2]*len(x)
y3 = [3]*len(x)

size = [2000]*len(x)  # size
```

### Plot

```python
line_width_val = 2

fig, ax = plt.subplots(
    figsize=(9,6),
    facecolor="white"
)
ax.set_title(
# fig.suptitle(
    "scatter plot with dense hatching",
    fontsize=20,
)

hatch_patterns = ["+", "++", "+++"]

# Plot scatter
for y, hatch_type in zip([y1, y2, y3], hatch_patterns):
    ax.scatter(
        x,y,
        size,
        color="white",
        edgecolor="black",
        linewidth=line_width_val,
        hatch=hatch_type,
        label=f"\"{hatch_type}\""
    )

# Set axis limit
ax.set_xlim(-0.5, 2.5)
ax.set_ylim(0.5, 3.5)

# Set axis labels
ax.set_xlabel(
    "x",
    fontsize=16
)
ax.set_ylabel(
    "y",
    fontsize=16
)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="both",
    which="major",
    left=False,
    labelleft=False,
    bottom=False,
    labelbottom=False
)

# Update spines thickness
for position in ["left", "top", "right", "bottom"]:
    ax.spines[position].set_linewidth(line_width_val)

ax.legend(
    bbox_to_anchor=(1.1, .5),
    loc="center",
    fontsize=16,
    frameon=False,
    markerscale=.4
)

plt.tight_layout()

plt.savefig(
    fname="plot5.png",
    format="png",
    dpi=125
);
```

## Mixed hatching

### Data

```python
y = [1, 1, 1]

x = range(len(y))
```

### Plot

```python
line_width_val = 2

fig, ax = plt.subplots(
    figsize=(9,6),
    facecolor="white"
)
fig.suptitle(
    "combining multiple hatch patterns",
    fontsize=20,
)

hatch_patterns = ["O.", "+o", "*-"]

# Plot bars
ax.bar(
    x=x,
    height=y,
    color="white",
    edgecolor="black",
    linewidth=line_width_val,
    hatch=hatch_patterns
)

# Set axis ticks and tick labels
ax.set_xticks(
    x,
    [f"\"{h}\"" for h in hatch_patterns],
    fontsize=14
)

# Set axis limit
# ax.set_ylim(0, 16)

# Remove spines
for position in ["left", "top", "right"]:
    ax.spines[position].set_visible(False)

# Remove `y` axis ticks and tick labels
ax.tick_params(
    axis="y",
    which="major",
    left=False,
    labelleft=False
)

# Update `x` axis tick thickness
ax.tick_params(
    axis="x",
    which="major",
    width=line_width_val,
    labelsize=20
)

ax.spines["bottom"].set_linewidth(line_width_val)

plt.tight_layout()

plt.savefig(
    fname="plot6.png",
    format="png",
    dpi=125
);
```
