## Word Cloud 2

```python
# Settings
width_px = 1200
height_px = 800
dpi = 100
```

```python
wordcloud = WordCloud(
    background_color="white",
    width=width_px,
    height=height_px,
    stopwords=[],
).generate(text)
```

```python
fig, ax = plt.subplots(
    figsize=(width_px/dpi, height_px/dpi),
    dpi=dpi
)
ax.imshow(wordcloud)
ax.axis("off");
```

## Word Cloud 3

```python
# Settings
width_px = 1200
height_px = 1200
dpi = 150
```

```python
# Read image & quick preview
img = Image.open("python_logo_bw.png")
plt.imshow(img);
```

```python
wordcloud = WordCloud(
    background_color="white",
    width=width_px,
    height=height_px,
    stopwords=[],
    mask=np.array(img),
    colormap="GnBu",
    font_path=Path("RubikDirt-Regular.ttf")
).generate(text)
```

```python
fig, ax = plt.subplots(
    figsize=(width_px/dpi, height_px/dpi),
    dpi=dpi,
)

ax.set_title(
    "Python 3.14 built-in functions",
    weight="bold",
    color="#03054d",
    fontsize=28,
    fontfamily="roboto"
)

ax.imshow(wordcloud)
ax.imshow(img, alpha=0.025)
ax.axis("off")

ax.text(
    1, -0.1,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#03054d",
    fontsize=10,
    family="monospace",
    weight="medium"
);
```
