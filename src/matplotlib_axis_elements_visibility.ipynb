# Difference between `ax.set_axis_off()` and `ax.set_frame_on()`

```python
import matplotlib.pyplot as plt
```

## 1. Default plot

```python
fig, ax = plt.subplots(
    facecolor="#f6f5f4",
    layout="constrained"
)
fig.suptitle("I'm an important Figure!")

ax.set_facecolor("#f9f06b")
ax.plot(
    [3, 1, 2, 1, 3],
    marker="o",
    lw=2,
    color="#e01b24"
)

ax.grid(
        color="#dc8add",
        lw=0.75
)
```

## 2. ax.set_axis_off()

```python
fig, ax = plt.subplots(
    facecolor="#f6f5f4",
    layout="constrained"
)
fig.suptitle("I'm an important Figure!")

ax.set_facecolor("#f9f06b")
ax.plot(
    [3, 1, 2, 1, 3],
    marker="o",
    lw=2,
    color="#e01b24"
)

ax.grid(
        color="#dc8add",
        lw=0.75
)

ax.set_axis_off()
```

## 3. ax.axis("off")

```python
fig, ax = plt.subplots(
    facecolor="#f6f5f4",
    layout="constrained"
)
fig.suptitle("I'm an important Figure!")

ax.set_facecolor("#f9f06b")
ax.plot(
    [3, 1, 2, 1, 3],
    marker="o",
    lw=2,
    color="#e01b24"
)

ax.grid(
        color="#dc8add",
        lw=0.75
)

ax.axis("off")
```

## 4. ax.set_frame_on(False)

```python
fig, ax = plt.subplots(
    facecolor="#f6f5f4",
    layout="constrained"
)
fig.suptitle("I'm an important Figure!")

ax.set_facecolor("#f9f06b")
ax.plot(
    [3, 1, 2, 1, 3],
    marker="o",
    lw=2,
    color="#e01b24"
)

ax.grid(
        color="#dc8add",
        lw=0.75
)

ax.set_frame_on(False)
```
