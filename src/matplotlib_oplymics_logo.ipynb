## Data

```python
logo_data = [
    ("blue", (-120, 0)),  # colour, xy for circle centre
    ("black", (0, 0)),
    ("red", (120, 0)),
    ("yellow", (-60, -50)),
    ("green", (60, -50)),
]
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22
    }
)
```

```python
fig, ax = plt.subplots(
    figsize=(16,8),
    facecolor="whitesmoke",
    layout="constrained"
)
fig.suptitle(
    "Olypics Logo",
    fontsize=40,
    weight="bold"
)

# Plot logo
for data in logo_data:
    artist = mpatches.Circle(
        xy=data[1],
        radius=50,
        facecolor=(0, 0, 0, 0),
        edgecolor=data[0],
        linewidth=20
    )
    ax.add_artist(artist)

ax.set_xlim(-200, 200)
ax.set_ylim(-120, 75)
ax.set_aspect("equal")
ax.set_axis_off()

# Footer
ax.text(
    1, 0,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=25,
    family="monospace",
    weight="medium"
)

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
