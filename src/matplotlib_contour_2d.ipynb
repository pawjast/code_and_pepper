## Data

```python
lim = 10
step = 0.1

X, Y = np.meshgrid(
    np.arange(-1.5*lim, 1.5*lim+step, step),
    np.arange(-lim, lim+step, step)
)

Z = 2*(X**2) + Y**2
```

## Settings

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "image.cmap": "YlGn",
    }
)
```

## Basic plot

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="white",
    layout="constrained"
)

ax.set_aspect("equal")
# fig.suptitle(
#     "TITLE",
#     fontsize=25,
#     weight="bold"
# )

con = ax.contourf(
    X, Y, Z,
)
cbar = fig.colorbar(con)

# Footer
ax.text(
    1, -0.13,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```

## Plot 2

```python
levels = np.concatenate(
    (
        np.arange(0, 120, 20),
        np.arange(200, 700, 100)
    )
)
```

```python
fig, ax = plt.subplots(
    figsize=(12, 8),
    facecolor="white",
    layout="constrained"
)
# fig.suptitle(
#     "TITLE",
#     fontsize=25,
#     weight="bold"
# )

# 2D Contour
con = ax.contourf(
    X, Y, Z,
    levels = levels
)
cbar = fig.colorbar(
    con,
    ticks=levels
)
# Contour lines
con = ax.contour(
    X, Y, Z,
    levels = levels,
    # colors="black"
    colors="#3d3846"
)
# Contour labels
ax.clabel(
    con
)

# Footer
ax.text(
    1, -0.13,
    "by @pawjast\nmade with @matplotlib \u2665",
    transform = ax.transAxes,
    horizontalalignment="right",
    verticalalignment="bottom",
    color="#9a9996",
    fontsize=16,
    family="monospace",
    weight="medium"
);

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
