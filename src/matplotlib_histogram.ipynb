# `matplotlib` histogram


## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
```

## Data

Generate fake distribution for height and weight.

```python
# DATA
nr = 2000  # sample size
data_height = [
    random.normalvariate(mu=176,sigma=4)
    for _ in range(nr)
]
data_weight = [
    random.normalvariate(mu=90,sigma=3)
    for _ in range(nr)
]
```

## Plot histogram of heights

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white"
)
fig.suptitle(
    "Average height",
    fontsize=24,
)

ax.hist(
    data_height,
    edgecolor="#613583",  # GNOME purple 5,
    facecolor ="#c061cb"  # GNOME purple 2
)

# labels
ax.set_xlabel (
    "bins",
    fontsize=16
)
ax.set_ylabel(
    "count",
    fontsize=16
)

# Update font size of the 'x' and 'y' axis labels
ax.tick_params(
    axis="both",
    labelsize=16
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot1.png",
    format="png",
    dpi=125
);
```

## Plot histogram of heights with custom bins - return n, bins, and patches and use it

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white"
)
fig.suptitle(
    "Average height",
    fontsize=24,
)

n, bins, patches = ax.hist(
    data_height,
    bins=15,
    edgecolor="#613583",  # GNOME purple 5,
    facecolor ="#c061cb"  # GNOME purple 2
)

# labels
ax.set_xlabel (
    "bins",
    fontsize=16
)
ax.set_ylabel(
    "count",
    fontsize=16
)
ax.bar_label(
    patches,
    fontsize=16,
    color="#613583",  # GNOME purple 5,
)

# set ticks
ax.set_xticks(bins)
ax.set_xticklabels([f"{value:.1f}" for value in bins])

# Update font size of the 'x' and 'y' axis labels
ax.tick_params(
    axis="both",
    labelsize=16
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot3.png",
    format="png",
    dpi=125
);
```

## Normalized

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white"
)
fig.suptitle(
    "Average height",
    fontsize=24,
)

n, bins, patches = ax.hist(
    data_height,
    bins=10,
    density=True,
    edgecolor="#613583",  # GNOME purple 5,
    facecolor ="#c061cb"  # GNOME purple 2
)

# labels
ax.set_xlabel (
    "bins",
    fontsize=16
)
ax.set_ylabel(
    "count",
    fontsize=16
)
ax.bar_label(
    patches,
    fontsize=16,
    color="#613583",  # GNOME purple 5
    fmt="{:,.3f}"
)

# set ticks
ax.set_xticks(bins)
ax.set_xticklabels([f"{value:.1f}" for value in bins])

# Update font size of the 'x' and 'y' axis labels
ax.tick_params(
    axis="both",
    labelsize=16
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot4.png",
    format="png",
    dpi=125
);
```

## Cummulative

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white"
)
fig.suptitle(
    "Average height",
    fontsize=24,
)

n, bins, patches = ax.hist(
    data_height,
    bins=10,
    cumulative=True,
    edgecolor="#613583",  # GNOME purple 5,
    facecolor ="#c061cb"  # GNOME purple 2
)

# labels
ax.set_xlabel (
    "bins",
    fontsize=16
)
ax.set_ylabel(
    "count",
    fontsize=16
)
ax.bar_label(
    patches,
    fontsize=16,
    color="#613583",  # GNOME purple 5
)

# set ticks
ax.set_xticks(bins)
ax.set_xticklabels([f"{value:.1f}" for value in bins])

# Update font size of the 'x' and 'y' axis labels
ax.tick_params(
    axis="both",
    labelsize=16
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot5.png",
    format="png",
    dpi=125
)
```

## 2D histogram

```python
fig, ax = plt.subplots(
    figsize=(16,12),
    facecolor="white"
)
fig.suptitle(
    "Weight - height relationship [MADE UP]",
    fontsize=24,
)

h, xedges, yedges, img = ax.hist2d(
    data_height,
    data_weight,
    bins=[12, 8],
    edgecolor="white",
    cmap="plasma"
)
fig.colorbar(img)

# set ticks
ax.set_xticks(xedges)
ax.set_xticklabels([f"{value:.1f}" for value in xedges])
ax.set_yticks(yedges)
ax.set_yticklabels([f"{value:.1f}" for value in yedges])

# labels
ax.set_xlabel (
    "heights",
    fontsize=16
)
ax.set_ylabel(
    "weights",
    fontsize=16
)

# Update font size of the 'x' and 'y' axis labels
ax.tick_params(
    axis="both",
    labelsize=16
)

# Remove spines
for position in ["top", "right"]:
    ax.spines[position].set_visible(False)

plt.tight_layout()

plt.savefig(
    fname="plot8.png",
    format="png",
    dpi=125
);
```
