# `matplotlib` plot a horn shape in 3D 

## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from matplotlib import font_manager

# Add system fonts for Matplotlib to see them
for font in font_manager.findSystemFonts(fontpaths=None, fontext='ttf'):
    font_manager.fontManager.addfont(font)
```

## Data

```python
nr_of_points = 200

x = np.linspace(0.75, 20, nr_of_points)  # Avoid x=0 to prevent division by zero
theta = np.linspace(0, 2 * np.pi, nr_of_points)
X, THETA = np.meshgrid(x, theta)

# Calculate r, y, and z
r = 1/X
Y = r * np.sin(THETA)
Z = r * np.cos(THETA)
```

## Plot

```python
plt.rcParams.update(
    {
        "font.family": "Roboto",
        "font.size": 22,
        "hatch.linewidth": 1.75,
        "axes.spines.top": False,
        "axes.spines.right": False,
    }
)
```

```python
fig = plt.figure(
    figsize=(10, 10),
    facecolor="#f6f5f4",
    layout="constrained"
)


ax = fig.add_subplot(
    projection="3d"
)
ax.set_aspect("equal")


# fig, ax = plt.subplots(
#     figsize=(16,12),
#     facecolor="white",
#     layout="constrained"
# )
fig.suptitle(
    "Horn shape",
    fontsize=40,
    weight="bold"
)

ax.plot_surface(
    X, Y, Z,
    color="#ff7800",  # orange
    alpha=0.65,
    edgecolor="#dc8add"
)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# plt.savefig(
#     fname="name.png",
#     format="png",
#     dpi=125,
#     metadata=metadata_dict
# );
```
